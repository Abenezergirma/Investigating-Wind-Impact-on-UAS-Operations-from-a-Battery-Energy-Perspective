```python
x_estimated_2 = 27.4 
x_std_2 = 0.4
x_axis_2 = np.linspace(x_estimated_2 - 3 * x_std_2, x_estimated_2 + 3 * x_std_2, 100)
y_axis_2 = norm.pdf(x_axis_2, x_estimated_2, x_std_2)
plt.plot(x_axis_2,y_axis_2, color="blue")

fit_pdf = norm.pdf(x_axis_2, x_estimated_2, x_std_2)

samples = np.random.choice(fit_pdf, size=30, p=fit_pdf/np.sum(fit_pdf)) 

first, second = [],[]
for i in range(len(samples)):
    location = np.where(fit_pdf == samples[i])[0]
    loc_1 = location[0]
    loc_2 = location[1]
    
    first.append(x_axis_2[loc_1])
    second.append(x_axis_2[loc_2])
    
plt.plot(first,samples,linestyle='', marker='s', color="blue")
plt.plot(second,samples,linestyle='', marker='s', color="blue")
plt.plot(x_axis_2,y_axis_2, color="red")
plt.title('Sampling from initial PDF (v)')
plt.savefig('Sampling_Points_from_PDFs', format='png')
```

```python
sampleTime = 0.01
Iapp = np.loadtxt("Iapp")
t = np.arange(0, 100, sampleTime)
t = np.round(t,2)
dicts = {}
loading = dict(zip(t,Iapp))

def tarrot_loading(t, x={}):
    
    if t < (len(loading)*sampleTime -0.01 ): 
    
        i = loading[round(t,2)]
    elif t < 100:
        i = 100
    else: 
        i =300
    return i
```

```python
current = []
for i in t:
#     print(i)
    current.append(tarrot_loading(i))
```

```python
plt.plot(t,current,linewidth=2, color='royalblue')
plt.xlabel('time')
plt.ylabel('current')
plt.title('Battery future loading (Iapp)')
plt.savefig('future_loading', format='png', dpi=140)
```
